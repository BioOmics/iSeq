```r
library(ggplot2)
library(patchwork)
```

```r
theme_chy <- function() {theme(
      axis.text = element_text(size = 12, colour = "black"),
      axis.line = element_line(colour = "black"),
      title = element_text(size = 15, colour = "black"),
      legend.text = element_text(size = 12, colour = "black"),
      legend.title = element_text(size = 15, colour = "black"),
      panel.background = element_blank()
    )
}
```

```r
timeENA1 <- read.table("./TestiSeq-540MB/timeENA.log", header = F, sep = "\t")
options(repr.plot.width = 4.5, repr.plot.height = 5)
ggplot(timeENA1, aes(x = V1, y = V2, color = V1)) +
  geom_boxplot(alpha = 0.8) +
  geom_jitter(position = position_jitter(width = 0.2), alpha = 0.7, size = 3) +
  scale_color_manual(values = c("Wget" = "#fc4e07", "Aspera" = "#00afbb", "AXEL" = "#e7b800")) +
  theme_chy() + theme(legend.position = "") +
  theme(axis.text = element_text(size = 15, colour = "black"))+
  labs(title = "FTP from ENA", x = "", y = "Time (s)")-> p1
p1
```

```r
timefasterqDump <- read.table("./TestFasterqDump/testFasterqDump.log", header = F, sep = "\t")
options(repr.plot.width = 5, repr.plot.height = 8)
ggplot(timefasterqDump, aes(x = V1, y = V3, fill = V2)) +
  geom_bar(stat = "identity", position = "stack", alpha = 0.8) +
  geom_smooth(aes(group = 1), method = "loess", se = FALSE, color = "black",linetype=2) +
  labs(x = "Threads", y = "Time (s)", fill = "") +
  scale_fill_manual(values = c("Convert SRA to FASTQ" = "#00afbb", "Compress FASTQ" = "#fc4e07")) +
  theme_chy() + 
  scale_x_continuous(limits = c(1,42), breaks = seq(2,42,4),expand = c(0,0))+
  scale_y_continuous(limits = c(0,900), breaks = seq(0,900,300),expand = c(0,0))+
  theme(axis.text = element_text(size = 15, colour = "black"), legend.position = c(0.7, 0.7))+
  coord_flip() -> p6
p6
```
